```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Page config
st.set_page_config(page_title="Superstore Dashboard", layout="wide")
st.title("ð Superstore Sales Dashboard")

# Load data
df = pd.read_csv(r"C:\Users\sumay\Desktop\SUPERSTORE PROJECT\Superstore_Orders_q1.csv")

# Clean column names
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Convert types
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')

# Handle missing values
df['Profit'] = df['Profit'].fillna(df['Profit'].median())
df['Discount'] = df['Discount'].fillna(df['Discount'].mean())

# Convert dates
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Feature engineering
df['Month'] = df['Order_Date'].dt.month

# Sidebar filters
st.sidebar.header("ð Filters")
region = st.sidebar.multiselect("Select Region", df['Region'].unique(), default=df['Region'].unique())
category = st.sidebar.multiselect("Select Category", df['Category'].unique(), default=df['Category'].unique())
filtered_df = df[(df['Region'].isin(region)) & (df['Category'].isin(category))]

# KPIs
st.subheader("ð Key Metrics")
col1, col2, col3 = st.columns(3)
col1.metric("Total Sales", f"{filtered_df['Sales'].sum():,.0f}")
col2.metric("Total Profit", f"{filtered_df['Profit'].sum():,.0f}")
col3.metric("Avg Discount", f"{filtered_df['Discount'].mean():.2f}")

# Charts
st.subheader("ð Visual Analysis")
col1, col2 = st.columns(2)

# Sales by Category
with col1:
    fig, ax = plt.subplots()
    sns.barplot(x='Category', y='Sales', data=filtered_df, ax=ax)
    ax.set_title("Sales by Category")
    st.pyplot(fig)

# Profit by Category
with col2:
    fig, ax = plt.subplots()
    sns.barplot(x='Category', y='Profit', data=filtered_df, ax=ax)
    ax.set_title("Profit by Category")
    st.pyplot(fig)

# Monthly trend
st.subheader("ð Monthly Sales Trend")

monthly = filtered_df.groupby('Month')['Sales'].sum()

fig, ax = plt.subplots()
sns.lineplot(x=monthly.index, y=monthly.values, marker='o', ax=ax)
ax.set_title("Monthly Sales")
st.pyplot(fig)

# Correlation heatmap
st.subheader("ð¥ Correlation")

fig, ax = plt.subplots()
sns.heatmap(filtered_df[['Sales','Profit','Discount']].corr(), annot=True, ax=ax)
st.pyplot(fig)
```
